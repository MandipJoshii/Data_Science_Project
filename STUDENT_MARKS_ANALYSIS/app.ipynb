```python
from dash import Dash, html, dcc, Input, Output
import pandas as pd
import plotly.express as px

app = Dash(__name__)

df = pd.read_csv("student-mat.csv")

app.layout = html.Div([
    html.H1("Student Final Grade Analysis Dashboard",
            style={'textAlign': 'center', 'color': '#2c3e50'}),

    html.Label("Select feature to explore vs G3:"),
    dcc.Dropdown(
        id="feature-dropdown",
        options=[
            {"label": "Study Time", "value": "studytime"},
            {"label": "Absences",   "value": "absences"},
            {"label": "G1 - First Period Grade", "value": "G1"},
            {"label": "G2 - Second Period Grade", "value": "G2"},
            {"label": "Failures",   "value": "failures"},
        ],
        value="G2"
    ),

    dcc.Graph(id="scatter-graph"),
    dcc.Graph(id="histogram-graph")
])

@app.callback(
    Output("scatter-graph", "figure"),
    Output("histogram-graph", "figure"),
    Input("feature-dropdown", "value")
)
def update_graphs(selected_feature):
    scatter = px.scatter(df, x=selected_feature, y='G3',
                         title=f"{selected_feature} vs Final Grade (G3)")
    hist = px.histogram(df, x=selected_feature,
                        title=f"Distribution of {selected_feature}")
    return scatter, hist

if __name__ == "__main__":
    app.run(debug=True)
```
