#### marker simply represents a line where x and y meets read the below simple example to understand it

#### few symbols are: o, ., v, d, s, p, * etc

```python
import matplotlib.pyplot as plt

y = [10,20,30]
x = [1,2,3]

plt.plot(x,y,marker='o', markersize='10', markerfacecolor='#a81fb3',markeredgecolor='#a81fb3') 

"""i can choose markersize = 10 or ms = 10 both does the same job ms is just the shorthand of markersize"""

"""note markerfacecolor can take color name like red, green, blue or hexa decimal code too """

"""and again we also have shorthand for markerfacecolor which is mfc = 'red' """

"""we can also color the edge of marker using markeregdecolor and its shorthand is mec = 'red' """
plt.show()
```
