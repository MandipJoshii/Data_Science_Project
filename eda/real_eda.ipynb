```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

student = pd.read_csv('../student-mat.csv',delimiter=';')

df = pd.DataFrame(student)  

df.head()
```

```python
xx = df['G1']
yy = df['G3']

plt.scatter(xx,yy)

plt.show()
```

```python
df.columns
```

### this is a visual exploration not the final statistical conclusion

```python
sample_df1 = df.sample(15)

x1 = sample_df1['absences']
y1 = sample_df1['G3']
plt.xlabel('student absence')
plt.ylabel('student G3')
plt.title("effect on G3 due to absences")
plt.scatter(x1,y1)

plt.show()
```

#### conclusion: from the exploratory scatter plot analysis between student absences and final exam grade (G3), no strong linear relationship is observed. While some students with fewer absences achieve higher grades, there are also cases where students with low attendance still perform well, and some with good attendance perform poorly. This indicates that absences alone are not a sufficient predictor of student performance and should be analyzed together with other academic and social factors in later stages of the project.

#### not final but for now the conclusion:

Scatter plots revealed that variables such as absences and study time show weak and inconsistent relationships with final grades, indicating that they are not strong standalone predictors. In contrast, previous academic performance (G2) demonstrated a strong positive relationship with G3, highlighting its importance for future predictive modeling. These findings confirm the necessity of multivariate analysis and motivate the use of machine learning models in subsequent stages of the project. At this stage, no predictive model has been implemented, and the analysis remains exploratory.

### NOTE: More bins = splitting data into many small intervals. Less bins = splitting data into fewer but larger intervals. But the total data represented is always the same — they're just different ways of viewing the same thing. More bins isn't always better, and less bins isn't always worse.
