```python
import numpy as np
import pandas as pd

student = pd.read_csv("../../student-mat.csv",delimiter=';')
df = pd.DataFrame(student)
```

```python
X = df['G2'].values
Y = df['G3'].values

X = (X-X.mean())/X.std()
Y = (Y-Y.mean())/Y.std()
```

```python
weight = 0.0
bias = 0.0
iteration = 1000
lr = 0.01
n = len(X)
```

```python


for i in range(iteration):

    # Forward pass
    y_pred = weight * X + bias

    
    loss = (1/n) * np.sum((Y - y_pred) ** 2)

    # Compute gradients
    dw = (-2/n) * np.sum(X * (Y - y_pred))
    db = (-2/n) * np.sum(Y - y_pred)

    
    weight -= lr * dw
    bias -= lr * db

    
    if i % 100 == 0:
        print(f"Iteration {i:4d} → Loss = {loss:.4f}  w = {weight:.4f}  b = {bias:.4f}")

print(f"\nFinal weight: {weight:.4f}")
print(f"Final bias:   {bias:.4f}")
print(f"Final loss:   {loss:.4f}")
```
